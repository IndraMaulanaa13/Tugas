```python
def soal1(n):
    sisi_kiri = sum([2*i-1 for i in range(1,n+1)])
    sisi_kanan = n**2
    return sisi_kiri == sisi_kanan
```

```python
soal1(5)
```

```python
def soal2(n):
    sisi_kiri = sum([2**i for i in range(0, n+1)])
    sisi_kanan = 2**(n+1)-1
    return sisi_kiri == sisi_kanan
```

```python
soal2(5)
```
